```python
import pandas as pd
from nltk.tokenize import word_tokenize
from collections import Counter

# Sample DataFrame with sentences


# Function to calculate term frequency
def calculate_term_frequency(text):
    # Tokenize the text into words
    words = word_tokenize(text.lower())  # Convert to lower case for uniformity
    total_words = len(words)
    
    # Calculate term frequency using Counter
    term_count = Counter(words)
    
    # Calculate TF for each term
    tf = {term: count / total_words for term, count in term_count.items()}
    
    return tf

# Apply the function to each row in the DataFrame
headlines['term_frequency'] = headlines['head_line'].apply(calculate_term_frequency)

# Display the DataFrame with term frequencies
print(headlines[['head_line', 'term_frequency']])
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from collections import Counter

# Sample DataFrame with sentences
data = {
    'text': [
        'This is the first document.',
        'This document is the second document.',
        'And this is the third one.',
        'Is this the first document?'
    ]
}

df = pd.DataFrame(data)

# Function to calculate term frequency
def calculate_term_frequency(text):
    # Tokenize the text into words
    words = word_tokenize(text.lower())  # Convert to lower case for uniformity
    total_words = len(words)
    
    # Calculate term frequency using Counter
    term_count = Counter(words)
    
    # Calculate TF for each term
    tf = {term: count / total_words for term, count in term_count.items()}
    
    return tf

# Apply the function to each row in the DataFrame
df['term_frequency'] = df['text'].apply(calculate_term_frequency)

# Display the DataFrame with term frequencies
print(df[['text', 'term_frequency']])
```

```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download stop words and punkt tokenizer if you haven't
nltk.download('stopwords')
nltk.download('punkt')

# Define stop words
stop_words = set(stopwords.words('english'))

# Sample text
text = "This is an example sentence to demonstrate stop stop stop stop stop stop stop word removal for word cloud generation."

# Tokenize and filter out stop words
words = word_tokenize(text)
filtered_sentence = [word for word in words if word.lower() not in stop_words]

# Convert filtered sentence back to a single string
filtered_text = ' '.join(filtered_sentence)

# Create the word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
